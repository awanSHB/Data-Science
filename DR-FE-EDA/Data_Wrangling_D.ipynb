```python
demographicsA = pd.read_csv("D:\PROGRAMMING\DATASETS\Korean_demographics_2000-2022.csv")
demographicsA
```

```python
print(demographicsA.shape)
print("rows->", demographicsA.shape[0])
print("columns->", demographicsA.shape[1])
```

```python
##Displaying the maximum of 5 column names in a 2d numpy array
cols = demographicsA.columns.to_list()
cols_split = [cols[i:i+5] for i in range(0, len(cols), 5)]
for i in cols_split:
    print(i)
```

```python
# 1) Checking the no of missing values for the each attribute
na_counts = demographicsA.isna().sum().to_dict()
na_counts_sorted = sorted(na_counts.items(),key = lambda x : x[1],  reverse=True)
for k in na_counts_sorted:
    print(k[0], ":", k[1])
```

```python
# 2) The above process will convert the dictionary into tuple, so to remain it in dict
na_counts = demographicsA.isna().sum().to_dict()
na_counts_sorted = {k:v for k,v in sorted(na_counts.items(), key=lambda x:x[1], reverse=True)}
for key, value in na_counts_sorted.items():
    print(key, ':', value)
```

```python
# 3) we can also do the above procedure without using the lambda function
na_counts = demographicsA.isna().sum().to_dict()
def sort(key):
    return key[1]
na_counts_sorted = {k:v for k,v in sorted(na_counts.items(), key=sort, reverse=True)}
for k,v in na_counts_sorted.items():
    print(k, ':', v)
```

```python
##Droping the missing values
demographicsB = demographicsA.dropna()
demographicsB.isna().sum()
```

```python
print(demographicsB.shape)
print(demographicsA.shape)
print("The Dropped Observatios are: ", demographicsA.shape[0]-demographicsB.shape[0])
```

```python
#checking for duplicated values
d = demographicsB[demographicsB.duplicated()]
d   #So there are no duplicat values
```

```python
#Check for the occurences of each observation in a certain column
countA = demographicsB['Region'].value_counts()
print(countA)
```

```python
#Grouping
group = demographicsB.groupby('Region')
groupA = group.get_group('Busan')
groupA[['Date','Marriage' ,'Divorce', 'Birth', 'Death']].head(10)
```

```python
#Finding out the unique values in column
v1 = demographicsB['Region'].unique()
print("There are ", v1.size, " different regions:")
for v in v1:
    print(v)
```

```python
demographicsB.loc[:, 'Successful_Marriages'] = demographicsB['Marriage']-demographicsB['Divorce']
demographicsB.loc[:, 'Successful_Marriage_Rate'] = demographicsB['Marriage_rate']-demographicsB['Divorce_rate']
len_a = len(demographicsB)         #Length before dropping rows
demographicsB
```

```python
#Here we are grouping the dataset by year 2000
#First we are dropping a whole country observation value
demographicsB = demographicsB.drop(demographicsB[demographicsB['Region']=='Whole country'].index)
len_b = len(demographicsB)          #Length After Dropping Rows
grouping = demographicsB.groupby(demographicsB['Date'].str[-4:])
demographicsC = grouping.get_group('2000')
print("Total Dropped Rows are : ", len_a-len_b)
demographicsC
```

```python
#We are finding the mean for the all regions in year 2000
#reset Index will reset the index to default sequential numeric index
demographicsC['Date'] = pd.to_datetime(demographicsC['Date'])
group_region_2000 = demographicsC.groupby('Region').mean().round(3).reset_index()
group_region_2000
```

```python
'''
--Here we are finding the average of each region in each year
'''
demographicsB['Date'] = pd.to_datetime(demographicsB['Date'])
demographicsB['Year'] = demographicsB['Date'].dt.year
demographicsD = demographicsB.groupby(['Region','Year']).mean().round(2).reset_index()
demographicsD
```

```python
demographicsD['Region'].value_counts()
```

```python
Check = demographicsD.groupby('Year').get_group(2000)
Check = Check.drop('Year', axis=1)
Check.head()
```

```python
demographicsE = demographicsD.copy()
demographicsE.drop(['Year', 'Date'], axis=1, inplace=True)
demographicsE = demographicsE.groupby('Region').mean().astype(int).reset_index()
demographicsE
```
