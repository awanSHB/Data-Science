```python
import pandas as pd
import numpy as np
import seaborn as sb
sb.set(style = 'darkgrid')
sb.set(rc={'figure.figsize': (12, 7)})
```

```python
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [10, 20, 30, 40, 50],
    'Year': [2015, 2016, 2017, 2018, 2019],
    'Gender': ['M', 'F', 'M', 'F', 'M'],
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Green'],
    'Height': [170, 165, 180, 175, 185],
    'Weight': [60, 70, 80, 90, 100],
    'Age': [20, 25, 30, 35, 40],
    'Score': [np.nan, 80, 70, 90, 75]
})

data
```

```python
sb.lineplot(x = 'Age', y = 'Height', data=data)
```

```python
sb.lineplot(x = 'Age', y = 'Height', data=data, hue = 'Gender')
```

```python
sb.scatterplot(x = 'Age', y = 'Weight', data=data)
```

```python
sb.scatterplot(x='Age', y = 'Weight', data=data, hue='Gender')
```

```python
sb.boxplot(data['Value'])
```

```python
sb.boxplot(x = 'Value', y = 'Height', data=data)
```

```python
sb.violinplot(data['Score'], orient='h')
```

```python
sb.pointplot(x = 'Age', y = 'Height', data=data)
```

```python
sb.pointplot(x = 'Age', y = 'Height', data=data, hue='Gender')
```

```python
sb.countplot(x = 'Gender', data=data)
```

```python
sb.countplot(x = 'Color', data=data)
```
