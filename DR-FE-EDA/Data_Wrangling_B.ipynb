```python
import pandas as pd
import numpy as np
```

```python
datasetA = pd.DataFrame(
    {
    'Name' : ['A', 'B', 'C', 'D', 'A', 'F', 'G', 'H', 'I', 'J', 'K', np.nan, 'B', 'N', 'O'],
    'Age'  : [10, 11, 12, 13, 10, np.nan, 15, 12, 13, 18, 19, 16, 15, 16, 20],
    'Marks': [71, 81, 72, 82, 72, 76, 77, np.nan, 88, 87, 86, np.nan, 85, 70, 73],
    'R_No' : [1, 2, 3, 4, 5, 6, 7, np.nan, 9, 10, 11, np.nan, 13, 14, 15]
    }
)
datasetA
```

```python
# print(datasetA.info())             #Index & datatypes
print(datasetA.dtypes)             #Datatype of each attribute
print(datasetA.empty)              #True for empty data frames
```

```python
print(datasetA.size)                #Rows * Colummns
print(datasetA.shape)               #(Rows, Columns)
print(datasetA.values)              #return a numpy array of dataframe
```

```python
print(datasetA.head())
print(datasetA.tail())
```

```python
datasetB = datasetA.dropna()                   #It will drop the columns with missing values
datasetB
```

```python
print(datasetB[datasetB.duplicated('Name')])         #display the duplicated values
datasetD = datasetB.drop_duplicates(subset='Name')
datasetD
```

```python
#Display those students whose marks are less than certain value
print(datasetD.loc[datasetD['Marks'] > 75])
print()
#Display Those Students whose marks are between certain values
print(datasetD.query('70 <= Marks <= 80'))
print()
print(datasetD[datasetD['Marks'].between(70, 80)])
```

```python
##Adding a new column
datasetD = datasetD.copy()
datasetD.loc[:, 'Percentage'] = ((datasetD['Marks']/90)*100).round(3)
datasetD.loc[:, 'D_O_B'] = 2023-datasetD['Age'].astype(int)
datasetD.loc[:, 'Deducted'] = 90-datasetD['Marks']
# datasetD.pop('Marks_Deducted')
datasetD
```

```python
#Sorting descending data by percentage
datasetF = datasetD.sort_values('Percentage', ascending=False)
datasetF
```
