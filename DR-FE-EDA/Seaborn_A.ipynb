```python
import numpy as np
import seaborn as sns
import pandas
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (19, 7)
```

```python
data = pandas.read_csv("D:\\PROGRAMMING\\DATASETS\\nba.csv")
data
```

```python
sns.lineplot(x = 'Age', y = 'Weight', data=data)
```

```python
sns.lineplot(x = 'Age', y = 'Weight', data=data, hue=data["Position"])
```

```python
dfA = pandas.DataFrame(
    {'Height': [65, 68, 70, 72, 63, 66, 69, 71],
    'Weight': [120, 140, 160, 180, 115, 135, 155, 175],
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Female', 'Female', 'Female', 'Female']})
sns.lineplot(x = 'Height', y = 'Weight', data=dfA, hue='Gender')
```
