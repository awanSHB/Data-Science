```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (14, 8)
```

```python
days = [ 'Saturday' , 'Sunday' , 'Monday' , 'Tuesday' , 'Wednesday' , 'Thursday' , 'Friday' ]
calories = [ 1670 , 2011 , 1853 , 2557 , 1390 , 2118 , 2063 ]

dfA = pd.DataFrame(
    {'Days': days, 'Calories': calories}
)

dfA.plot('Days', 'Calories')
plt.show()
```

    Two variables at y-axis 

```python
subjects = [ 'Math' , 'English' , 'History' , 'Chem' , 'Geo' , 'Physics' , 'Bio' , 'CS' ]
# Create a list of data to be 
# represented in y-axis
stress = [ 9 , 3 , 5 , 1 , 8 , 5 , 10 , 2 ]
  
# Create second list of data
# to be represented in y-axis
grades = [ 15 , 10 , 7 , 8 , 11 , 8 , 17 , 20 ]

dfB = pd.DataFrame(list(zip(stress, grades)), index=subjects, columns=['Stress', 'Grades'])
dfB
```

```python
dfB.plot(marker = 'o')
plt.show()
```

```python
subjects = [ 'Math' , 'English' , 'History' , 'Chem' , 'Geo' , 'Physics' , 'Bio' , 'CS' ]

stress = [ 9 , 3 , 5 , 1 , 8 , 5 , 10 , 2 ]

grades = [ 15 , 10 , 7 , 8 , 11 , 8 , 17 , 20 ]

dfB_A = pd.DataFrame({'Subjects': subjects, 'Stress': stress, 'Grades': grades})
ax = dfB_A.plot(x='Subjects', y='Stress', label='Stress', marker='o')
dfB_A.plot(x='Subjects', y='Grades', ax=ax, label='Grades', marker='s')

# Create the legend
legend = plt.legend()

# Change the legend text color to red
for text in legend.get_texts():
    text.set_color('black')
plt.show()
```

```python
c = 30

dfC = pd.DataFrame([c + k for k in range(0, 30)],
                   index=[*range(1, 31)],
                   columns=['Temperature (C)'])
dfC
```

```python
ax = dfC.plot(title='Line Plot')
legend = plt.legend()

# Change the legend text color to black
for text in legend.get_texts():
    text.set_color('black')

plt.xlabel('Index')
plt.ylabel('Temperature (C)')
plt.title('Line Plot')
plt.show()
```
