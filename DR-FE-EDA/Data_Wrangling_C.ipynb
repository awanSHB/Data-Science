```python
import pandas as pd
import numpy as np
```

```python
datasetA = pd.DataFrame(
    {
    'Name' : ['A', 'B', 'C', 'D', 'A', 'F', 'G', 'H', 'I', 'J', 'K', np.nan, 'B', 'N', 'O'],
    'Age'  : [10, 11, 12, 13, 10, np.nan, 15, 12, 13, 18, 19, 16, 15, 16, 20],
    'Marks': [71, 81, 72, 82, 72, 76, 77, np.nan, 88, 87, 86, np.nan, 85, 70, 73],
    'R_No' : [1, 2, 3, 4, 5, 6, 7, np.nan, 9, 10, 11, np.nan, 13, 14, 15]
    }
)
datasetA
```

```python
datasetB = pd.DataFrame(
    {
    'Name' : ['A', 'B', 'C', 'D', 'A', 'F', np.nan, 'H', 'I', 'J', 'K', 'L', 'B', 'N', 'O'],
    'Gender': ['M', 'F', 'M', 'M', 'F', 'M','M', 'F', 'M','M', 'F', 'M','M', 'F', 'M'],
    'Major': ['CS', 'SE', 'DS', 'AI', 'CS', 'SE', 'DS', 'AI', 'CS', 'SE', 'DS', 'AI', 'CS', 'SE', 'DS']
    }
)
datasetB
```

```python
datasetC = pd.DataFrame(pd.merge(datasetA, datasetB, on="Name"))
datasetC
```

```python
##Dealing with the missing values
datasetC['Marks'] = datasetC['Marks'].fillna(datasetC['Marks'].mean())
datasetC
```

```python
#Removing the missing values
datasetC = datasetC.dropna()
datasetC
```

```python
##Dealing with duplicates
datasetC = datasetC.drop_duplicates('Name')
datasetC
```

```python
print(datasetC.shape)       #(Rows, Columns)
print(datasetC.shape[0])    #Rows
print(datasetC.shape[1])    #columns
```

```python
def findGrade(marks):
    if(marks >= 50 and marks <= 60):
        return 'C'
    elif(marks >60 and marks<=70):
        return 'B'
    elif(marks > 70 and marks <= 80):
        return 'A'
    elif(marks > 80):
        return 'A+'
    else:
        return 'F'
```

```python
datasetC.loc[:, 'Percentage'] = ((datasetC['Marks']/90)*100).round(2)
datasetC.loc[:, 'Grade'] = datasetC['Marks'].apply(findGrade)
datasetC.loc[:, 'Gender'] = datasetC['Gender'].replace({'M': 1, 'F': 0})
datasetC
```

```python
grouping = datasetC.groupby('Major')
grouping.get_group('DS')
```

```python
#Select Those students whose marks are between 70 and 84, display theri name, gender, rno, major
datasetC.query('70<= Marks <= 84')[['Name', 'R_No', 'Gender', 'Major']]
```
