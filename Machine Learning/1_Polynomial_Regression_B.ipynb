```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (7, 4)
```

```python
df = pd.read_csv("D:\PROGRAMMING\Datasets\polydata.csv")
df
```

```python
## Independent / dependent variables
X = df.iloc[:, 1:2]
y = df.iloc[:, 2]
```

```python
def lin_reg_model():
    lin = LinearRegression()
    lin.fit(X, y)
    return lin.predict(X)
```

```python
def train_model(deg):
    polreg = PolynomialFeatures(degree=deg)
    x_poly = polreg.fit_transform(X)
    linreg = LinearRegression()
    linreg.fit(x_poly, y)
    y_pred = linreg.predict(x_poly)
    return y_pred
```

```python
def visualize(y_axis, t):
    plt.scatter(X, y, color = "blue")
    plt.plot(X, y_axis, color = "red")
    plt.title(t)
    plt.xlabel("Position Level")
    plt.ylabel("Salary")
    return plt.show()
```

```python
plt.style.use('fivethirtyeight')
y_pred_lin = lin_reg_model()
y_pred_poly = train_model(2)
print("rmse : ", np.sqrt(mean_squared_error(y, y_pred_poly)))
visualize(y_pred_lin, "Linear Regression")
visualize(y_pred_poly, "Polynomial Regression (Deg = 2)")
```

```python
y_pred_poly = train_model(3)
print("rmse : ", np.sqrt(mean_squared_error(y, y_pred_poly)))
visualize(y_pred_poly, "Polynomial Regression (Deg = 3)")
```

```python
y_pred_poly = train_model(4)
print("rmse : ", np.sqrt(mean_squared_error(y, y_pred_poly)))
visualize(y_pred_poly, "Polynomial Regression (Deg = 4)")
```

```python
y_pred_poly = train_model(5)
print("rmse : ", np.sqrt(mean_squared_error(y, y_pred_poly)))
visualize(y_pred_poly, "Polynomial Regression (Deg = 5)")
```
